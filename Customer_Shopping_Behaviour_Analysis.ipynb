```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
```

```python
df.info()
```

```python
df = df.rename(columns={'purchase_amount_(usd)': 'purchase_amount_usd'})
```

```python
df.info()
```

```python
#Create a column age_groups
labels=['Young', 'Adults', 'Middle_Aged', 'Senior']
df['age_group']=pd.qcut(df['age'],q=4, labels=labels)
```

```python
df[['age','age_group']].head(10)
```

```python
#Purchase frequency days
df['frequency_of_purchases'].unique()
```

```python
frequency_mapping = {
    'Fortnightly': 14, 
    'Weekly' : 7, 
    'Annually' : 365, 
    'Quarterly' : 90, 
    'Bi-Weekly' : 14,
    'Monthly' : 30, 
    'Every 3 Months' : 90
}

df['purchase_frequency_days']= df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['frequency_of_purchases', 'purchase_frequency_days']].head(10)
```

```python
df[['discount_applied', 'promo_code_used']].head(10)
```

```python
(df['discount_applied']==df['promo_code_used']).all()
```

```python
df = df.drop('promo_code_used', axis=1)
```

```python
from sqlalchemy import create_engine
import urllib
```

```python
params = urllib.parse.quote_plus(
    "DRIVER={ODBC Driver 17 for SQL Server};"
    "SERVER=PRINCE-S;"
    "DATABASE=customer_behaviour;"
    "Trusted_Connection=yes;"
)

engine = create_engine(f"mssql+pyodbc:///?odbc_connect={params}")
```

```python
pd.read_sql("SELECT name FROM sys.databases", engine)
```

```python
df.to_sql(
    name="customer_shopping_behaviour_cleaned",
    con=engine,
    if_exists="replace",
    index=False
)
```
